Visualisation Plots

```python
# Function for Bar Chart - Coffee Distribution

def plot_coffee_distribution(data):

    """
    Plots a bar chart of coffee type distribution.

    Parameters:
    data (DataFrame): The dataset containing transaction information with at least
                      a 'coffee_name' column.

    This function creates a bar chart to show the number of purchases for each
    coffee type, giving insights into popular items.
    """

    # Setting the figure size
    plt.figure(figsize=(10, 6))

    # Calculate the count of each coffee type using value_counts()
    coffee_counts = data['coffee_name'].value_counts()

    coffee_counts.plot(kind='bar', color='skyblue')

    # Adding a title
    plt.title("Distribution of Coffee Types Purchased")

    plt.xlabel("Coffee Type")

    plt.ylabel("Number of Purchases")

    # Rotating x-axis labels by 45 degrees
    plt.xticks(rotation=45)

    # Displaying plot
    plt.show()
```

```python
# Function for Money Spent Scatter plot

def plot_coffee_cash_scatter(data):

    """
    A scatter plot with coffee name on the y-axis,
    cash type as hue with distinct colors,
    and money spent on the x-axis.

    Parameters:
    data (DataFrame): The data containing 'cash_type', 'money',
                      and 'coffee_name' columns.

    Returns:
    None
    """

    # Setting up figure size
    plt.figure(figsize=(12, 8))

    # Creating scatter plot
    sns.scatterplot(
        data=data,
        x='money',
        y='coffee_name',
        hue='cash_type',
        palette={"card": "blue", "cash": "orange"},
        s=100,
        alpha=0.7
    )

    # Labelling the x-axis as 'Money Spent'
    plt.xlabel('Money Spent')

    # Labelling the y-axis as 'Coffee Type'
    plt.ylabel('Coffee Type')

    # Set the title for the plot
    plt.title('Scatter Plot of Coffee Type by Money Spent with Cash Type')

    # Setting legend title as 'Cash Type'
    plt.legend(title='Cash Type')

    # Adding grid lines
    plt.grid(True)

    # Show the plot
    plt.show()
```

```python
# Function for Box Plot - money spent by coffee type

def plot_money_by_coffee_type(data):

    """
    a box plot for the distribution of money spent by coffee type.

    Parameters:
    data (DataFrame): The dataset containing transaction information with at least
                      'coffee_name' and 'money' columns.

    This function will create a box plot to display the distribution of transaction
    amounts for each coffee type, allowing insights into spending patterns by
    coffee variety.
    """

    # Setting figure size
    plt.figure(figsize=(12, 6))

    # Create a box plot with coffee types on the x-axis and transaction amounts on the y-axis
    sns.boxplot(x='coffee_name', y='money', data=data, palette='Pastel1')

    # Setting title of the plot
    plt.title("Transaction Amount by Coffee Type")

    plt.xlabel("Coffee Type")

    plt.ylabel("Transaction Amount (Money)")

    # Rotating the x-axis labels by 45 degrees for readability
    plt.xticks(rotation=45)

    # Displaying plot
    plt.show()

```

Function Calling

```python
# Plot the coffee type distribution
plot_coffee_distribution(data)

# plot for money spent
plot_coffee_cash_scatter(data)

# Calling function to plot and analyze transaction amounts by coffee type
plot_money_by_coffee_type(data)
```
